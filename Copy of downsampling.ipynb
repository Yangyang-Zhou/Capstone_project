# Policy id level

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
id_train = pd.read_csv('/content/drive/MyDrive/BU Capstone/Results/Before midterm/Final sample design results/id_Xtrain.csv').drop('Unnamed: 0', axis=1)
id_train.head()
```

```python
# seperate hit and no-hit
hit = id_train[id_train.y == 0]
no_hit = id_train[id_train.y == 1]
```

```python
# downsampling
hit_down = resample(hit, replace=False, n_samples=(len(no_hit)*2), random_state=2, stratify=hit[['STATE','year']])
```

```python
# combine downsampled hit and no-hit
id_down = pd.concat([hit_down, no_hit])
id_down.y.value_counts()
```

```python
id_down.to_csv('id_downsample.csv')
```
